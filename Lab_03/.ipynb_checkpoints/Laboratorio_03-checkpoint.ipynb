```python
# EROSION
kernel = np.array([[1,1,1],
                   [1,1,1],
                   [1,1,1]], dtype=np.uint8)

def erosion(img, kernel):
    height, width = img.shape
    k_height, k_width = kernel.shape
    pad_height, pad_width = k_height // 2, k_width // 2
    print(f"Imagen:{height}-{width}  Kernel:{k_height}-{k_width}  Pad:{pad_height}-{pad_width}")

    # Crear una imagen de salida con el mismo tamao que la imagen de entrada
    output_img = np.zeros((height, width), dtype=np.uint8)

    # Recorrer cada pxel de la imagen
    for i in range(pad_height, height-pad_height):
        for j in range(pad_width, width-pad_width):
            print(i, j)
            # Aplicar el kernel al vecindario del pxel
            neighborhood = img[i-pad_height:i+pad_height+1, j-pad_width:j+pad_width+1]
            #print(i-pad_height, i+pad_height+1, j-pad_width, j+pad_width+1)
            print(neighborhood)
            #result = (neighborhood == kernel).all()
            #output_img[i, j] = 255 if result else 0

    return output_img

```
