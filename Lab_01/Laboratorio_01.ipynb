```python
import os
import cv2 #OpenCV
import numpy as np #Biblioteca Numerica
import matplotlib.pyplot as plt #Visualizaciones

from zipfile import ZipFile
from urllib.request import urlretrieve

from IPython.display import Image

%matplotlib inline
```

```python
#Carga una imagen a color.
image = cv2.imread("coca-cola.png")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
#Extrae el canal Red de la imagen y almacenalo en una matriz. 
print(image.shape)
y = image.shape[0]
x = image.shape[1]
image_gray = np.full((y, x), 0)

for i in range(0,y):
    for j in range(0,x):
        image_gray[i][j] = image[i][j][2]

print(image_gray)
plt.imshow(image_gray, cmap="gray")
```

```python
#Obtener el negativo de la imagen..
image_gray_reverse = np.copy(image_gray)
for i in range(0,y):
    for j in range(0,x):
        image_gray_reverse[i][j] = 255 - image_gray_reverse[i][j]


print(image_gray_reverse)
plt.imshow(image_gray_reverse, cmap="gray")
```

```python
#Reflejar verticalmente la imagen original..
image_reflected = np.copy(image_gray)
print(y)
print(x)
print(image_reflected.shape)

for i in range(y):
    for j in range(x):
        # Intercambiar los valores de los pxeles verticalmente
        image_reflected[i][j] = image_gray[y-i-1][j]


print(image_reflected)
plt.imshow(image_reflected, cmap="gray")
```
